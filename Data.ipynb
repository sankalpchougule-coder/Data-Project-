```python
df.columns = ["ym","exp_imp","country","customs","hs9_code","Q1","Q2","Value"        
]
```

```python
df.columns
```

```python
cat_col = df.select_dtypes(include = ['object']).columns.tolist()
num_cols = df.select_dtypes(include = ['int64', 'float64']).columns.tolist()

print("Categorical columns:", cat_col)
print("Numeric columns:", num_cols)

for col in cat_col:
    print(f"\nValue counts for '{col}':")
    print(df[col].value_counts())
```

```python
df['country'].value_counts()
```

```python
missing_df = df.isnull().sum().reset_index()
missing_df.columns = ['column', 'Missing_Values']
missing_df['Missing_Percent'] = (missing_df['Missing_Values'] / len(df)) * 100
missing_df.sort_values('Missing_Percent', ascending=False)
```

```python
monthly = df.groupby('ym')['Value'].sum()

plt.figure(figsize=(10,5))
plt.plot(monthly.index, monthly.values)
plt.title("Monthly Trend of Total Trade Value")
plt.xlabel("Period (ym)")
plt.ylabel("Total Trade Value")
plt.show()
```

```python
country_counts = df['country'].value_counts().head(10)

plt.figure(figsize=(10,5))
country_counts.plot(kind='bar', color='green')
plt.title("Top 10 Countries by Frequency")
plt.xlabel("Country Code")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()


```

```python
exp_counts = df['exp_imp'].value_counts()

plt.figure(figsize=(6,4))
exp_counts.plot(kind='bar', color='skyblue')
plt.title("Value Count: Import vs Export")
plt.xlabel("exp_imp")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

```python
exp_counts = df['exp_imp'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(exp_counts, labels=exp_counts.index, autopct='%1.1f%%')
plt.title("Import vs Export Distribution")
plt.show()
```
